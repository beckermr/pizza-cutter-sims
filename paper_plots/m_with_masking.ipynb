```python
def _process_data(data):
    s2n_cuts = []
    mfrac_cuts = []
    mns = []
    stds = []

    for line in data.splitlines():
        if line.startswith("["):
            continue

        if line.startswith("cuts"):
            items = line.split(":", 1)[1].strip().split(",")
            s2n_cuts.append(int(items[0].split(">=")[1]))
            mfrac_cuts.append(int(items[2].split(":")[1].strip()))

        if line.strip().startswith("noise cancel m"):
            items = line.split(":")[1].split("[")[0].split("+/-")
            mns.append(float(items[0]))
            stds.append(float(items[1]))

    d = np.zeros(len(s2n_cuts), dtype=[
        ("s2n", "i8"), ("mfrac", "i8"), ("mn", "f8"), ("std", "f8")
    ])
    d["s2n"] = s2n_cuts
    d["mfrac"] = mfrac_cuts
    d["mn"] = mns
    d["std"] = stds
    
    return d
```
